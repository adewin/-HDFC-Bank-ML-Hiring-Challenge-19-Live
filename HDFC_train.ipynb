```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
```

```
df_train=pd.read_csv("DataSet/Train.csv")
df_test=pd.read_csv("DataSet/Test.csv")
```

```
#remove those columns that have only Nan values
for i in range(3,2395):
    try:
        p=all(df_train['Col{}'.format(i)].isnull())
        if p :
            df_train=df_train.drop(['Col{}'.format(i)],axis=1)
    except KeyError:
        pass


```

```
#preprocessing 
#1.check dataset contains any null
#2.check categorical data
#3.need decimal scalling

df_train.isnull().any(axis=0)
```

```
for i in range(3,2398):
    try:
      #calculate sum of nan values of every columns and go ahead
        sum_Nan=df_train['Col{}'.format(i)].isnull().sum()
        #70% data nan then remove columns
        #
        x70=(17521*7)/10
        x50=(17521*5)/10
        #if Nan value found then if condition true
        if sum_Nan >=x70:
            df_train=df_train.drop(['Col{}'.format(i)],axis=1)
           #if columns have morethan 50% data nan value and less 70% then  direct fill with 0 
        elif  sum_Nan <x70 and sum_Nan>x50:
            df_train['Col{}'.format(i)]=df_train['Col{}'.format(i)].fillna(np.mean(0))

        elif sum_Nan<x50:
            #then fill mean values
            df_train['Col{}'.format(i)]=df_train['Col{}'.format(i)].fillna(np.mean(df_train['Col{}'.format(i)]))


    except KeyError:
        pass
```

```
#extract featrues and labels
labels=df_train['Col2'].values        
features=df_train.iloc[:,2:]

#now perform decimal scalling
from sklearn.preprocessing import StandardScaler   
sc=StandardScaler()

features=sc.fit_transform(features)

```

```
from sklearn.model_selection import train_test_split
features_train,features_test,labels_train,labels_test=train_test_split(features,labels,random_state=0,test_size=0.2)
```

```
from sklearn.linear_model import  LogisticRegression
lr=LogisticRegression()
lr.fit(features_train,labels_train)
pred_lr=lr.predict(features_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm_lr=confusion_matrix(pred_lr,labels_test)
print(cm_lr)
print("Accuracy of Logistic =",accuracy_score(labels_test,pred_lr))
```

```
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(features_train,labels_train)
pred_dt=dt.predict(features_test)
cm_dt=confusion_matrix(pred_dt,labels_test)
print(cm_dt)
print("Accuracy of DecisionTree =",accuracy_score(labels_test,pred_dt))
```

```
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
gnb=GaussianNB()
bnb=BernoulliNB()
mnb=MultinomialNB()
gnb.fit(features_train,labels_train)
pred_gnb=gnb.predict(features_test)
cm_gnb=confusion_matrix(pred_gnb,labels_test)
print(cm_gnb)
print(accuracy_score(labels_test,pred_gnb))
```

```
bnb.fit(features_train,labels_train)
pred_bnb=bnb.predict(features_test)
cm_bnb=confusion_matrix(pred_bnb,labels_test)
print(cm_bnb)
print(accuracy_score(labels_test,pred_bnb))
```

```
#apply knn algo
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,p=2)
knn.fit(features_train,labels_train)
pred_knn=knn.predict(features_test)
cm_pred=confusion_matrix(pred_knn,labels_test)
print(cm_pred)
print(accuracy_score(labels_test,pred_knn))
```

```
from sklearn.svm import  SVC
from sklearn.ensemble import RandomForestClassifier
svc=SVC(kernel='rbf',degree=3)
svc.fit(features_train,labels_train)
pred_svc=svc.predict(features_test)
cm_pred=confusion_matrix(pred_svc,labels_test)
print(accuracy_score(labels_test,pred_svc))
```

```
rf= RandomForestClassifier(n_estimators=20, random_state=0)  
rf.fit(features_train,labels_train)
pred_rf=rf.predict(features_test)
print(confusion_matrix(pred_rf,labels_test))
print(accuracy_score(pred_rf,labels_test))
```
