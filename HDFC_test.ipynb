```
#remove those columns that have only Nan values
def remove_nan(df):
  for i in range(3,2395):
      try:
          p=all(df['Col{}'.format(i)].isnull())
          if p :
              df=df.drop(['Col{}'.format(i)],axis=1)
      except KeyError:
          pass
  return df

```

```
def cleaning_rows(df):
    list1=[]
    for i in range(3,2398):
        try:
          #calculate sum of nan values of every columns and go ahead
            sum_Nan=df['Col{}'.format(i)].isnull().sum()
            #70% data nan then remove columns
            #
            x70=(17521*7)/10
            x50=(17521*5)/10
            #if Nan value found then if condition true
            if sum_Nan >=x70:
                #print(i)
                list1.append(i)
                df=df.drop(['Col{}'.format(i)],axis=1)
              #if columns have morethan 50% data nan value and less 70% then  direct fill with 0 
            elif  sum_Nan <x70 and sum_Nan>x50:
                df['Col{}'.format(i)]=df['Col{}'.format(i)].fillna(np.mean(0))

            elif sum_Nan<x50:
                #then fill mean values
                df['Col{}'.format(i)]=df['Col{}'.format(i)].fillna(np.mean(df['Col{}'.format(i)]))


        except KeyError:
            pass
    return df,list1
```

```
df_train=remove_nan(df_train)
df_train,list1=cleaning_rows(df_train)
```

```
df_test=remove_nan(df_test)
df_test,list2=cleaning_rows(df_test)
```

```
#extract those columns that removed in test file and not yet removed from train files
extra_columns=[x for x in list2 if x not in list1]
```

```
def remove_extra_columns(df,extra_columns):
  for i in extra_columns:
    df=df.drop(['Col{}'.format(i)],axis=1)
  return df
```
